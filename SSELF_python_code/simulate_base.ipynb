```python
#simulate_base.ipynb
#Testing without hierarchy to make sure the base works
import time
import random
import numpy as np
from SSELF_base import System, FootprintDatabase

# Step 1: Initialize the system and footprint DB
system = System(num_companies=10, num_products=10)
footprint_db_2024 = FootprintDatabase(year=2024)

# Step 1: Initialize footprint database with 0 values
for company in system.companies.values():
    for product in company.products:
        footprint_db_2024.report(product.product_id, 0)

system.solve(footprint_db_2024, forced_updates=0, verbose=True)
```


# Step 3: Iterative update loop
updates_completed = False
iteration = 0
start_time = time.time()

while not updates_completed:
    iteration += 1
    print(f"\n--- Iteration {iteration} ---")
    any_company_updated = False

    companies_to_check = list(system.companies.keys())
    random.shuffle(companies_to_check)

    for cname in companies_to_check:
        company = system.companies[cname]
        print(f"Checking {cname}")
        if company.check_update_needed(footprint_db_2024):
            print(f"  → Updated. New: {company.latest_update:.4f}")
            any_company_updated = True
        else:
            print("  → No update needed.")

    updates_completed = not any_company_updated

end_time = time.time()

# Step 4: Summary
print("\n✅ Updates completed.")
print(f"⏱️ Time taken: {end_time - start_time:.2f} seconds")

print("\n📦 Final Footprints:")
for cname, company in system.companies.items():
    print(f"{cname}: {company.latest_update:.4f} kg CO2e/unit")

print("\n🔁 Update Count:")
total_updates = sum(c.num_updates for c in system.companies.values())
for cname, c in system.companies.items():
    print(f"{cname}: {c.num_updates} updates")
print(f"Total updates: {total_updates}")
